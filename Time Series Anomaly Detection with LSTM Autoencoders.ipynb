## Background

### Anomaly Detection

Anomaly detection, also known as outlier detection or novelty detection, is the task of identifying observations that are unusual or do not conform to the expected behavior within a given dataset. Anomaly detection has a wide range of applications in various fields, and is an important task in data analysis as it allows the identification and handling of unusual observations in a dataset. This can be useful in several ways:

* First, anomaly detection can be used as a component of a data cleaning pipeline to identify and remove outliers that may be errors or noise in data collection or processing. 

* Second, anomaly detection can identify rare events of interest, such as cyber-attacks or fraudulent transactions, that may not be easily detectable using other methods. 

Thus, detecting and analyzing anomalies provides valuable insights into the behaviour of a system so that appropriate measures to mitigate any potential risks or issues can be taken.

#### Time Series Anomalies

Time series anomalies, also known as temporal anomalies, refer to unusual events or patterns within a time series data. Time series data is a sequence of observations collected over a period and can be found in a wide range of applications, including finance, weather forecasting, and the stock market. These anomalies can be either global or local, persistent or transient, and can vary in their degree of anomalousness.

Time series anomaly detection involves analyzing the temporal dependencies within the data and identifying deviations from the expected behavior.

image.jpg

There are several approaches to anomaly detection, including statistical, machine learning, and deep learning techniques. Statistical methods involve analyzing the statistical properties of the data and identifying observations that deviate significantly from the expected distribution. Machine learning approaches involve training a model on a normal training dataset and then using the model to identify deviations from the expected behavior in the test dataset. This project examines the deep learning approach of training a neural network to reconstruct the normal data and then using it to identify deviations from the normal data distribution.

### Long Short-Term Memory (LSTM) RNNs

Recurrent neural networks (RNNs) are a type of neural network that is well-suited for processing sequential data, such as time series or natural language text. RNNs contain feedback connections, which allow the network to process data temporally and retain information from previous time steps. This makes RNNs particularly effective at modelling dependencies between the elements of a sequence.

One limitation of RNNs is their inability to effectively handle long-term dependencies, as the information from the earlier time steps may become diluted as it is passed through the recurrent connections. To address this issue, long short-term memory (LSTM) networks were introduced. LSTM networks are a variant of RNNs that use additional gates to control the flow of information through the network, allowing them to effectively model long-term dependencies.

LSTM networks have been successfully applied to a variety of tasks, including language modelling, machine translation, and time series forecasting. They have also been used for anomaly detection in time series data, as they can model the temporal dependencies within the data and identify deviations from the expected behaviour.

### Autoencoder Models

Autoencoders are a type of neural network used for unsupervised learning. They consist of an encoder-decoder (sequence-to-sequence) architecture, and are trained to reconstruct the input data as accurately as possible. The encoder maps the input data to a lower-dimensional latent space, and the decoder maps the latent representation back to the original input space. During training, the autoencoder minimizes the reconstruction error between the input data and the reconstructed output.



image.jpg

One of the main advantages of autoencoders is their ability to learn the underlying structure of the data and identify patterns that may not be easily detectable using other methods. Autoencoders can be applied to various tasks, including data denoising, dimensionality reduction, and anomaly detection.

There are several variations of autoencoders (and more generally, seq2seq architectures) including recurrent neural network (RNN) models, convolutional neural network (CNN) models, and transformer models.

### LSTM Autoencoders

LSTM autoencoders are a type of neural network that combines the capabilities of LSTM networks and autoencoders for the purpose of anomaly detection in time series data. LSTM autoencoders are trained to reconstruct the input time series data as accurately as possible, using the LSTM network as the encoder and decoder. Anomaly detection is then performed by comparing the reconstruction error of the input data to a predefined threshold.

Thus, an autoencoder can be used to perform anomaly detection by encoding a sequence of observations into a low-dimensional latent representation and then using this compressed representation to reconstruct the input sequence. Any deviations between the model's predictions and the actual values may indicate the presence of anomalies.

image (1).jpg

One advantage of using LSTM autoencoders for anomaly detection is their ability to model long-term dependencies within the data. This is important in many real-world time series datasets, as anomalies may not always be immediately apparent and may require information from multiple time steps to be detected. This makes LSTM autoencoders effective at detecting abnormalities in various types of time series data, including sensor, financial, and energy consumption data.

Additionally, this approach allows the use of the model's prediction errors as a measure of the likelihood that an anomaly is present. This can be useful for ranking anomalies by their severity or for setting thresholds for determining which anomalies to investigate further.

### Preprocess the Dataset

Before training and evaluating an LSTM autoencoder model on the time series data, it is necessary to preprocess the dataset. This involves several steps to prepare the data for modelling.

* The first step in preprocessing the dataset is to split it into three data frames: a training set, a validation set, and a test set. This is important to ensure that the model is trained and evaluated on separate datasets, which helps prevent overfitting and evaluate the generalizability of the model.

* The next step is to normalize the data. Normalization is scaling the data to have zero mean and unit variance, which helps stabilize the training process and improve the model's performance.

* Finally, the data needs to be reshaped to be compatible with an LSTM autoencoder model in Keras. This involves transforming the data into a 3D tensor with the shape (`samples`, `timesteps`, `features`), where `samples` is the number of time series sequences, `timesteps` is the length of each sequence, and `features` is the number of variables in each sequence. This reshaped data is then ready to be fed into an LSTM autoencoder model for training and evaluation.

#### Split the Dataset

There are several ways to split a time series dataset into training, validation, and test sets. The most common approaches are:

* Sequential split: This involves dividing the time series data into contiguous blocks, with the first block being the training set, the second block as the validation set, and the third block as the test set. This approach is straightforward but problematic if the time series data exhibits global temporal shift, as the validation and test sets may contain values from different distributions from the training set.

* Stratified split: This involves dividing the time series data into blocks based on the temporal dependencies in the data. For example, if the data exhibits daily seasonality, it may be divided into blocks corresponding to individual days. This approach helps to ensure that the validation and test sets are representative of the data distribution, but it can be more complex to implement and may require more data to be effective.

This sections sequentially splits the data. However, if the models's performance is not satisfactory, the stratified split might help improve it. 

* The training set consists of all samples with a timestamp earlier than `01/04/2014`, which is a period of approximately nine months.

* The validation set consists of all samples with a timestamp between `01/04/2014` and `01/05/2014` (one month).

* The test set consists of all samples with a timestamp later than `01/05/2014` (one month).

```python
# Set the start dates for the validation and test sets
val_start_date = "2014-04-01"
test_start_date = "2014-05-01"

# Split the dataframe
train_df = df[df["timestamp"] < val_start_date]
test_df = df[df["timestamp"] >= test_start_date].reset_index(drop=True)
val_df = df[(
    df["timestamp"] >= val_start_date) & (df["timestamp"] < test_start_date
    )].reset_index(drop=True)

# Print the number of samples in each dataframe 
print("Percentages of Samples:")
print(f"Training Set:\t{round(len(train_df) / len(df) * 100, 2)}%")
print(f"Validation Set:\t{round(len(val_df) / len(df) * 100, 2)}%")
print(f"Test Set:\t{round(len(test_df) / len(df) * 100, 2)}%")
```

#### Standardize the Dataset

Standardizing a dataset is the process of transforming the data so that it has zero mean and unit variance. This is often useful when training machine learning models, as it can help to stabilize the training process and improve the model's performance.

To standardize a dataset, the mean and standard deviation of each feature is first calculated. The data is then transformed by subtracting the mean and dividing it by the standard deviation of each feature. This results in a new dataset with zero mean and unit variance.

```python
# Instantiate a StandardScaler object
scaler = StandardScaler()

# Fit the scaler on the training set
scaler = scaler.fit(train_df[["temperature"]])

# Transform all three splits
pd.options.mode.chained_assignment = None
train_df["temperature"] = scaler.transform(train_df[["temperature"]])
test_df["temperature"] = scaler.transform(test_df[["temperature"]])
val_df["temperature"] = scaler.transform(val_df[["temperature"]])

train_df.head()
```

#### Reshape the Dataset

To fit an LSTM autoencoder model in Keras, the time series data needs to be reshaped into a 3D tensor with the shape (`samples`, `timesteps`, `features`). `samples` refers to the number of time series sequences, `timesteps` refers to the length of each sequence, and `features` refers to the number of variables in each sequence.

```python
# Hyperparameter: timesteps in each sequence
timesteps = 4

def to_sequences(x):
  """
  Converts time series data into sequences of a given length.

  Parameters:
    - x (pandas DataFrame): The input time series data.

  Returns:
    - tuple: The input sequences.
  """
  
  # Initialize a list to store the sequences
  x_values = []

  # Iterate over the input data
  for i in range(len(x) - timesteps):
      
      # Append a sequence of the desired length to the list of input sequences
      x_values.append(x.iloc[i:(i + timesteps)].values)
      
  # Convert the list to NumPy array and return it
  return np.array(x_values)

# Reshape the dataframes
x_train = to_sequences(train_df[["temperature"]])
x_val = to_sequences(val_df[["temperature"]])
x_test = to_sequences(test_df[["temperature"]])

print(f"x_train.shape = {x_train.shape}")
```

### Create and Fit the Model

This is an LSTM autoencoder for anomaly detection on a univariate time series. The model has an encoder-decoder architecture, where the encoder processes the input time series and produces a fixed-length low-dimensional encoding or representation of the input. The decoder takes this encoding and reconstructs the original time series as accurately as possible.

The input to the model has shape (`batch_size`, `timesteps`, `features`), where `batch_size` is the number of sequences in the batch, `timesteps` are the number of time steps in each sequence, and `features` is the number of `features` at each time step.

* The encoder is a single-layer LSTM with `latet_dim` hidden units, which means it produces a `latet_dim`-dimensional encoding of the input time series, which contains all the important information about all timesteps for each sample. The encoding is passed through a repeat vector layer, which repeats the encoding ``timesteps`` times to produce a 3D tensor of shape (`batch_size`, `timesteps`, `latet_dim`).

* The decoder is also a single-layer LSTM with `latet_dim` hidden units, and it takes the repeated encoding as input and tries to, together with the following Dense layer, reconstruct the original time series as accurately as possible. The decoder produces a 3D tensor of shape (`batch_size`, `timesteps`, `latent_dim`) as output. The Dense layer is applied to each timestep to produce the final prediction at that timestep. 

The model is compiled with the "MSE" loss function, so it's optimized to minimize the mean squared error between the model's reconstructions and the actual input values.

```python
# Hyperparameters:
features = 1
latent_dim = 16

# Define the input shape of the model
input_shape = (timesteps, features)

# Define the input layer
inputs = Input(shape=input_shape, name="input_layer")

# Define the encoder LSTM layer
lstm_encoder = LSTM(latent_dim, name="lstm_encoder")(inputs)

# Define the repeat vector layer
repeat_vector = RepeatVector(timesteps, name="repeat_vector")(lstm_encoder)

# Define the decoder LSTM layer
lstm_decoder = LSTM(latent_dim, return_sequences=True, name="lstm_decoder")(repeat_vector)

# Define the dense layer
dense = Dense(1, name="dense")(lstm_decoder)

# Define the model
model = Model(inputs=inputs, outputs=dense)

# Compile the model
model.compile(optimizer="adam", loss="mse")

# Print the model summary
model.summary()
```

```python
model.fit(x_train, x_train, epochs=8, batch_size=256, validation_data=(x_val, x_val))
```

### Anomaly Detection

Once the model is trained, the anomaly detection process involves using the model to reconstruct the input time series as accurately as possible and comparing the model's reconstructiosn to the actual input values. Any deviations may indicate the presence of anomalies.

Once the deviations in the training set are measured, it is typically necessary to set a threshold for determining which are considered anomalies. This threshold can be determined through trial and error, or it can be set based on the characteristics of the data and the desired level of sensitivity.

```python
def predictions_mse(x, y):
  """
  Calculate the mean squared error (MSE) between the model's predictions 
  and the actual values for a given dataset.

  Parameters:
    - x: (pandas dataframe) The input data.
    - y: (pandas dataframe) The ground truth.

  Returns:
    - y_mse: (np.ndarray) The MSE between y_pred and y_true.
  """

  # Get the model's predictions
  y_pred = model.predict(x)

  # Calculate the MSE between the predictions and the actual values
  y_mse = (np.square(y - y_pred)).mean(axis=1)

  # Plot a histogram of the MSE values
  ax = sns.histplot(y_mse, bins=50)

  # Set the y-axis scale to logarithmic
  ax.set(title="MSE Histogram", xlabel="MSE", ylabel="Frequency", yscale="log")

  # Show the plot
  plt.show()

  return y_mse
```

```python
def prediction_error(mse_threshold, df_split, y_mse):
  """
  Calculate the prediction errors for a given split of data, 
  and plots them with a given mean squared error (MSE) threshold.

  Parameters:
    - mse_threshold: (float) The MSE threshold used to determine whether a prediction is considered an anomaly
    - df_split: The dataframe for the split
    - y_mse: The MSE values for the split

  Returns:
    - mse_df: A dataframe containing the prediction error data
  """

  # Create a dataframe from the split data
  mse_df = pd.DataFrame(df_split[timesteps:])

  # Add the MSE values to the dataframe
  mse_df["mse"] = y_mse

  # Add the MSE threshold to the dataframe
  mse_df["max_mse"] = mse_threshold

  # Add a column indicating whether the MSE is above the threshold (i.e., an anomaly)
  mse_df["anomaly"] = mse_df["mse"] > mse_df["max_mse"]

  # Add the temperature values to the dataframe
  mse_df["temperature"] = df_split[timesteps:]["temperature"]

  # Plot the MSE values and the MSE threshold
  plt.figure(figsize=(16, 8))
  ax = sns.lineplot(x=mse_df["timestamp"], y=mse_df["mse"])
  sns.lineplot(x=mse_df["timestamp"], y=mse_df["max_mse"])
  ax.set(title="MSE Time Series", xlabel="Timestamp", ylabel="MSE")

  return mse_df
```

```python
def plot_anomalies(mse_df):
  """
  Plot the anomalies in a given dataframe of mean squared error (MSE) values.

  Parameters:
    - mse_df: A dataframe containing the MSE values and anomaly data
  """

  # Filter the dataframe to include only rows where an anomaly was detected
  anomalies = mse_df[mse_df["anomaly"] == True]

  # Plot the temperature values and anomalies
  plt.figure(figsize=(16, 8))
  sns.lineplot(x=mse_df["timestamp"], y=mse_df["temperature"])
  ax = sns.scatterplot(x=anomalies["timestamp"], y=anomalies["temperature"], color="hotpink", marker="x")
  ax.set(title="Temperature Time Series", xlabel="Timestamp", ylabel="MSE")
  
  # Show the plot
  plt.show()
```
